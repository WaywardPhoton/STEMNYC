You know that applying AND, OR, and NOT gates (among others) to classical bits will *change* the information contained. There is a way to manipulate quantum information in a similar way, using *quantum gates* as part of a *quantum circuit*. 

There are three parts of a quantum circuit.

- Wires. Wires represent the program flow or time development of qubits.
- Gates. Gates apply transformations to qubits. 
- Measurements. This is the action of measuring the qubit in the $|0\rangle , |1\rangle $ basis.

![](images/qc.PNG)

Let's explore some of the gates that we are going to use as part of this activity. We will be looking at 

- X gates
- H gates
- Z gates
- I gates

...and using our knowledge of them to build circuits. 

Recall that the states $ |0\rangle $ and $|1\rangle$ can be written as column vectors. So quantum gates, which effect a transformation on these vectors, must have a matrix representation. 

1. The X gates is identical to the NOT gates that we saw in classical circuits. Applying a NOT gates to a state will  turn it into its opposite:

$ X(|0\rangle) = |1\rangle $ and $X(|1\rangle) = |0\rangle$.

The X gate can be written as $\begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix}$

2. The H gate, when applied to a normal state, will create a *superposition*:

$ H (|0\rangle) = \frac{1}{\sqrt{2}}|0\rangle +\frac{1}{\sqrt{2}}|1\rangle$ and $ H(|1\rangle) = \frac{1}{\sqrt{2}}|0\rangle -\frac{1}{\sqrt{2}}|1\rangle$ 

Note that the $\frac{1}{\sqrt{2}}|0\rangle +\frac{1}{\sqrt{2}}|1\rangle$ state is an importatnt state called the $|+\rangle$ state, and the $\frac{1}{\sqrt{2}}|0\rangle -\frac{1}{\sqrt{2}}|1\rangle$ state is called the $|-\rangle$ state. These states lie on the +x and -x axes of the Bloch sphere respectively. 

The H gate can be written as  $\frac{1}{\sqrt{2}}\begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix}$.

![](images/hgate.PNG)

3. The Z gate adds a phase shift to the $|1\rangle $ state:
$Z(|0\rangle) = |0\rangle$ and $Z(|1\rangle) = -|1\rangle$

The Z gate can be written as $\begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}$.

4. The I gate applied an identity matrix to the state and does not change it in any way. 

The I gate is written as  $\begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix}$.
